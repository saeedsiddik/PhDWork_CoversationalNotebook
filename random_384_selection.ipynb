```python
def random_selection(sample_size):
    
    # Sample size (tiny sample)
    num_nbname_to_select = sample_size


    # Seed for reproducibility  
    # Ref: https://arxiv.org/pdf/2109.08203

    random_seed = 3407

    with open(file_containing_all_kgtorrentdataName, 'r') as file:
        lines = file.readlines()

    if len(lines) < num_nbname_to_select:
        raise ValueError(f"The file only has {len(lines)} lines, which is less than {num_nbname_to_select}.")


    random.seed(random_seed)
    selected_lines = random.sample(lines, num_nbname_to_select)

    with open(file_containing_all_sampleName, 'a') as file:
        file.writelines(selected_lines)

    print(f"{num_nbname_to_select} lines have been randomly selected from {len(lines)} and written to {output_file} with seed {random_seed}.")
```

# get notebook cell info

```python
def read_code_cells_with_numbers(notebook_path):
    with open(notebook_path, 'r', encoding='utf-8') as f:
        notebook = json.load(f)

    code_cells = []

    for i, cell in enumerate(notebook['cells']):
        if cell['cell_type'] == 'code':
            # Combine the source lines into a single string for each code cell
            code = ''.join(cell['source'])
            # Store the cell number and code as a tuple
            code_cells.append((i+1, code))  # i+1 to make the cell number 1-based

    return code_cells
```

```python
# Filter out lines containing any of the specified strings
def update_sample_file(sample_file, strings_to_remove):
    with open(sample_file, 'r') as file:
        lines = file.readlines()

    # Filter lines: Keep only those that do not match any string in the removal list
    filtered_lines = []
    for line in lines:
        remove = False
        for s in strings_to_remove:
            if s in line:
                remove = True
                break  # Exit inner loop if any string matches
        if not remove:
            filtered_lines.append(line)
            
    
    updated_file = sample_file

    # Write the filtered lines to the output file
    with open(updated_file, 'w') as file:
        file.writelines(filtered_lines)
```

# copy the list of sample files from storage to local directory

```python


def copy_files(file_list, destination_dir):
    """
    Copies a list of files from the source directory to the destination directory.

    :param file_list: List of file names to copy.
    :param source_dir: Path to the source directory.
    :param destination_dir: Path to the destination directory.
    """
    
    number_of_file_to_copy = 0
    
    try:
        # Ensure the destination directory exists
        if not os.path.exists(destination_dir):
            os.makedirs(destination_dir)

        for file_name in file_list:
            
            nb_filename_only = file_name.split('/')[-1]
            source_path = file_name.replace(nb_filename_only, '')
            
            
            # source_path = os.path.join(source_path, nb_filename_only)
            source_path = file_name
            destination_path = os.path.join(destination_dir, nb_filename_only)
            
            # Check if the file exists in the source directory
            if os.path.exists(source_path):
                shutil.copy(source_path, destination_path)
                # print(f"Copied: {file_name}")
                number_of_file_to_copy += 1
            else:
                print(f"File not found: {file_name}")
                
        print(f"{number_of_file_to_copy} File copy operation completed.")
        
    except Exception as e:
        print(f"An error occurred: {e} in {file_name}")
```

```python
def correcting_filePath_list(filePath_list):
    nb_path_corrected_list = []
    
    for file_path in filePath_list:
#     make a corrected to match the path in storage 
        nb_path_corrected = file_path.replace('nb_data/', '')
        nb_path_corrected = nb_path_corrected.replace('\n', '')
        dataset_path_prefix = '/storage/saeed/nb_data_kgtorrent/'
        nb_path_corrected = dataset_path_prefix+nb_path_corrected
        
        nb_path_corrected_list.append(nb_path_corrected)
        
    return nb_path_corrected_list
```
